# REST request example

Esempio aggiornato per il modulo M3.

Obiettivi:
- fare una richiesta HTTP con `requests`
- leggere una risposta JSON
- estrarre campi semplici e campi annidati
- confrontare due endpoint della stessa API

```python
import requests

company_overview_url = 'https://www.alphavantage.co/query?function=OVERVIEW&symbol=IBM&apikey=8CLKWQB8AZ4LRY40'
company_indicators_url = 'https://www.alphavantage.co/query?function=INCOME_STATEMENT&symbol=IBM&apikey=8CLKWQB8AZ4LRY40'
```

```python
company_overview_url
```

```python
overview_response = requests.get(company_overview_url, timeout=30)
indicators_response = requests.get(company_indicators_url, timeout=30)

overview = overview_response.json()
indicators = indicators_response.json()

print('HTTP status overview:', overview_response.status_code)
print('HTTP status indicators:', indicators_response.status_code)
print('Company name:', overview.get('Name'))
print('Sector:', overview.get('Sector'))

annual_reports = indicators.get('annualReports', [])
if annual_reports:
    first_report = annual_reports[0]
    print('Latest fiscal date:', first_report.get('fiscalDateEnding'))
    print('Revenue:', first_report.get('totalRevenue'))
    print('Net income:', first_report.get('netIncome'))
else:
    print('No annual reports returned by the API.')
```
